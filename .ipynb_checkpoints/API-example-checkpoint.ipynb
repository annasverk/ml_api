## ml_api example

Что можно сделать с помощью __ml_api__?
1. Вернуть список доступных для обучения классов моделей
2. Обучить ML-модель с возможностью настройки гиперпараметров. При этом гиперпараметры для разных моделей могут быть разные. Количество классов моделей доступных для обучения = 4
3. Вернуть предсказание конкретной модели (система умеет хранить несколько обученных моделей)
4. Обучить заново и удалить уже обученные модели

Сперва необходимо запустить файл `setup.py` или `app.py`

```python
import requests
import io
import pandas as pd
from PIL import Image
from sklearn.datasets import load_iris, load_boston
```

Демонстрировать возможности API будем на датасетах _boston_ (задача регрессии) и _iris_ (задача классификации)

```python
iris_data = load_iris()
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris_df['species'] = iris_data.target
iris_json = iris_df.to_json()

boston_data = load_boston()
boston_df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
boston_df['PRICE'] = boston_data.target
boston_json = boston_df.to_json()
```

Метод __get__ возвращает список доступных для обучения моделей, а точнее алгоритмов из `sklearn`. Это сделано для того, чтобы сразу определялся тип задачи – регрессия или классификация. Всего реализовано 4 класса моделей:
- Линейные модели
- Метод опорных векторов
- Дерево решений
- Случайный лес

```python
resp = requests.get('http://0.0.0.0:5000/ml_api')
resp.json()
```

Чтобы обучить модель, необходимо вызвать метод __post__ и передать в параметр __json__ всю необходимую информацию, а именно:
- model (str) – модель, которую необходимо обучить (нужно выбрать из списка)
- data (json) – данные в формате json, на которых надо обучить модель (поддерживаются только количественные признаки!)
- params (dict) – гиперпараметры модели, по умолчанию используются дефолтные значения соответствующего алгоритма в `sklearn`
- grid_search (bool) – надо ли делать подбор оптимальных значений гиперпараметров, по умолчанию нет
- param_grid (dict) – сетка для перебора значений гиперпараметров, по умолчанию используется заданная сетка

```python
payload = {'model': 'RandomForestRegressor',
           'data': boston_json}
response = requests.post('http://0.0.0.0:5000/ml_api', json=payload)
response
```

Внутри данного метода, помимо сохранения обучающих и тестовых данных (разбиение производится автоматически случайным образом в соотношении 7:3), обучения модели и сохранения модели, также рассчитываются метрики качества в зависимости от класса модели. 

Чтобы посмотреть информацию обо всех обученных моделях, надо вызвать метод __get__ по адресу `./all_models`. Для каждой модели хранится ее id, название, гиперпараметры, качество на трейне и тесте, было ли обучение модели заново и была ли она удалена. Это своего рода файл истории использования API. Данные и модель сохраняются в файлы, в названии которых указан id модели, в папки data и models соответственно. 

```python
response = requests.get('http://0.0.0.0:5000/ml_api/all_models')
response.json()
```

Параметры модели, также как и сетку для их перебора, можно либо задать вручную, либо не задавать вовсе. В последнем случае используются дефолтные параметры. 

```python
payload = {'model': 'RandomForestRegressor',
           'params': {'n_estimators': 200, 'max_depth': 10, 'min_samples_leaf': 3},
           'data': boston_json}
response = requests.post('http://0.0.0.0:5000/ml_api', json=payload)
response
```

API поддерживает работу только с количественными признаками. Если попытаться подать категориальные данные, выпадет ошибка 400. 

```python
boston_df_cat = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
boston_df_cat['category'] = 'a'
boston_df_cat['PRICE'] = boston_data.target
boston_json_cat = boston_df_cat.to_json()
```

```python
payload = {'model': 'RandomForestRegressor',
           'data': boston_json_cat}
response = requests.post('http://0.0.0.0:5000/ml_api', json=payload)
response.status_code, response.json()
```

Также ошибка выпадет, если пытаться строить модель не из списка доступных; использовать модель классификации для решения задачи регрессии и наоборот; передавать неправильный гиперпараметр (не поддерживаемый данным алгоритмом)

```python
payload = {'model': 'RandomForestClassifier',
           'data': boston_json}
response = requests.post('http://0.0.0.0:5000/ml_api', json=payload)
response.status_code, response.json()
```

```python
payload = {'model': 'GradientBoostingRegressor',
           'data': boston_json}
response = requests.post('http://0.0.0.0:5000/ml_api', json=payload)
response.status_code, response.json()
```

```python
payload = {'model': 'RandomForestRegressor',
           'params': {'learning_rate': 0.1},
           'data': boston_json}
response = requests.post('http://0.0.0.0:5000/ml_api', json=payload)
response.status_code, response.json()
```

Вернуть предсказания конкретной модели на трейне и тесте можно с помощью метода __get__ с указанием в пути id модели. Например:

```python
response = requests.get('http://0.0.0.0:5000/ml_api/1')
response.json()
```

Под обучением заново понимается обучение модели на новых данных. Сделать это можно с помощью метода __put__. Необходимо в адресе запроса указать id модели, а в параметр __json__ передать новые данные (также в формате json). В этом случае в словаре со всеми моделями retrained станет True. 

```python
response = requests.put('http://0.0.0.0:5000/ml_api/1', json=boston_df.sample(100).to_json())
response
```

```python
response = requests.get('http://0.0.0.0:5000/ml_api/all_models')
response.json()['1']
```

Чтобы удалить уже обученную модель, надо вызвать метод __delete__ и указать в адресе id модели. При корректном выполнении выпадает ошибка 204, а deleted в словаре меняет свое значение на True.

```python
response = requests.delete('http://0.0.0.0:5000/ml_api/1')
response
```

```python
response = requests.get('http://0.0.0.0:5000/ml_api/all_models')
response.json()['1']
```

При попытке вызвать какой-либо метод для удаленной модели выпадает ошибка 404. Например:

```python
response = requests.get('http://0.0.0.0:5000/ml_api/1')
response.status_code, response.json()
```

В качестве усложнения данного API была реализована возможность рисовать графики для оценки качества обученной модели – график остатков для алгоритма регрессии и матрица ошибок для алгоритма классификации. Для этого нужно вызвать метод __get__ по адресу `./plot/id`, где id – это id модели, для которой хотим нарисовать график. 

```python
response = requests.get('http://0.0.0.0:5000/ml_api/plot/2')
img = Image.open(io.BytesIO(response.content)) 
img
```

```python
requests.post('http://0.0.0.0:5000/ml_api', json={'model': 'SVC', 'data': iris_json})
response = requests.get('http://0.0.0.0:5000/ml_api/plot/3')
img = Image.open(io.BytesIO(response.content)) 
img
```

Также для дерева решений и случайного леса можно вывести график важности признаков – адрес `./plot_feature_importance/id`. Для иных моделей будет выдаваться ошибка 400.

```python
response = requests.get('http://0.0.0.0:5000/ml_api/plot_feature_importance/2')
img = Image.open(io.BytesIO(response.content)) 
img
```

```python
response = requests.get('http://0.0.0.0:5000/ml_api/plot_feature_importance/3')
response.status_code, response.json()
```
